```python
# 1ë‹¨ê³„: NumPy ê¸°ì´ˆ - ììœ¨ì£¼í–‰ ì„¼ì„œ ë°ì´í„° ì²˜ë¦¬

# ì½”ë©ì—ì„œ ë°”ë¡œ ì‹¤í–‰ ê°€ëŠ¥!



import numpy as np

print("ğŸš— ììœ¨ì£¼í–‰ NumPy ê¸°ì´ˆ í•™ìŠµ ì‹œì‘!")

print("ğŸ“¦ NumPy ë²„ì „:", np.__version__)

print()



# ===== 1. NumPy ë°°ì—´ ê¸°ì´ˆ =====

print("=== 1. NumPy ë°°ì—´ ê¸°ì´ˆ ===")



# 1-1. ë¦¬ìŠ¤íŠ¸ë¥¼ NumPy ë°°ì—´ë¡œ ë³€í™˜

print("ğŸ“Š 1-1. ì„¼ì„œ ê±°ë¦¬ ë°ì´í„°")

sensor_distances = [8.5, 12.3, 15.7, 2.1, 9.8]  # íŒŒì´ì¬ ë¦¬ìŠ¤íŠ¸

print("ì›ë³¸ ë¦¬ìŠ¤íŠ¸:", sensor_distances)

print("ë¦¬ìŠ¤íŠ¸ íƒ€ì…:", type(sensor_distances))



# NumPy ë°°ì—´ë¡œ ë³€í™˜

np_distances = np.array(sensor_distances)

print("NumPy ë°°ì—´:", np_distances)

print("NumPy íƒ€ì…:", type(np_distances))

print("ë°°ì—´ ëª¨ì–‘:", np_distances.shape)

print("ë°°ì—´ í¬ê¸°:", np_distances.size)

print()



# 1-2. ì§ì ‘ ë°°ì—´ ìƒì„±

print("ğŸ“Š 1-2. ë‹¤ì–‘í•œ ë°°ì—´ ìƒì„± ë°©ë²•")

zeros_array = np.zeros(5)  # 0ìœ¼ë¡œ ì±„ìš´ ë°°ì—´

ones_array = np.ones(4)    # 1ë¡œ ì±„ìš´ ë°°ì—´

range_array = np.arange(0, 10, 2)  # 0ë¶€í„° 10ê¹Œì§€ 2ì”© ì¦ê°€

linspace_array = np.linspace(0, 100, 5)  # 0ë¶€í„° 100ê¹Œì§€ 5ê°œ êµ¬ê°„



print("0ìœ¼ë¡œ ì±„ìš´ ë°°ì—´:", zeros_array)

print("1ë¡œ ì±„ìš´ ë°°ì—´:", ones_array)

print("ë²”ìœ„ ë°°ì—´ (0~10, 2ì”©):", range_array)

print("êµ¬ê°„ ë°°ì—´ (0~100, 5ê°œ):", linspace_array)

print()



# ===== 2. ë°°ì—´ ì—°ì‚° =====

print("=== 2. ë°°ì—´ ì—°ì‚° ===")



# 2-1. ê¸°ë³¸ ì‚°ìˆ  ì—°ì‚°

print("ğŸ“Š 2-1. ì„¼ì„œ ë°ì´í„° ê³„ì‚°")

distances = np.array([8.5, 12.3, 15.7, 2.1, 9.8])

print("ì›ë³¸ ê±°ë¦¬:", distances)



# ëª¨ë“  ê°’ì— ë™ì¼í•œ ì—°ì‚° ì ìš©

distances_in_cm = distances * 100  # ë¯¸í„°ë¥¼ ì„¼í‹°ë¯¸í„°ë¡œ

distances_plus_offset = distances + 1.5  # ëª¨ë“  ê°’ì— 1.5 ë”í•˜ê¸°



print("ì„¼í‹°ë¯¸í„° ë³€í™˜:", distances_in_cm)

print("ì˜¤í”„ì…‹ ì ìš©:", distances_plus_offset)

print()



# 2-2. ë°°ì—´ ê°„ ì—°ì‚°

print("ğŸ“Š 2-2. ë‘ ì„¼ì„œ ë°ì´í„° ë¹„êµ")

front_sensors = np.array([8.5, 12.3, 15.7])

rear_sensors = np.array([6.2, 9.1, 13.4])



print("ì „ë°© ì„¼ì„œ:", front_sensors)

print("í›„ë°© ì„¼ì„œ:", rear_sensors)

print("ì„¼ì„œ ì°¨ì´:", front_sensors - rear_sensors)

print("í‰ê·  ê±°ë¦¬:", (front_sensors + rear_sensors) / 2)

print()



# ===== 3. ë°°ì—´ í†µê³„ í•¨ìˆ˜ =====

print("=== 3. ë°°ì—´ í†µê³„ í•¨ìˆ˜ ===")



print("ğŸ“Š 3-1. ì„¼ì„œ ë°ì´í„° í†µê³„")

all_distances = np.array([8.5, 12.3, 15.7, 2.1, 9.8, 6.2, 11.4, 3.9])

print("ì „ì²´ ì„¼ì„œ ë°ì´í„°:", all_distances)



print("ìµœëŒ€ê°’:", np.max(all_distances))

print("ìµœì†Œê°’:", np.min(all_distances))

print("í‰ê· ê°’:", np.mean(all_distances))

print("ì¤‘ê°„ê°’:", np.median(all_distances))

print("í‘œì¤€í¸ì°¨:", np.std(all_distances))

print("í•©ê³„:", np.sum(all_distances))

print()



# 3-2. ì¸ë±ìŠ¤ ì°¾ê¸°

print("ğŸ“Š 3-2. íŠ¹ì • ê°’ì˜ ìœ„ì¹˜ ì°¾ê¸°")

max_index = np.argmax(all_distances)  # ìµœëŒ€ê°’ì˜ ì¸ë±ìŠ¤

min_index = np.argmin(all_distances)  # ìµœì†Œê°’ì˜ ì¸ë±ìŠ¤



print(f"ê°€ì¥ ë¨¼ ê±°ë¦¬: {all_distances[max_index]}m (ì¸ë±ìŠ¤: {max_index})")

print(f"ê°€ì¥ ê°€ê¹Œìš´ ê±°ë¦¬: {all_distances[min_index]}m (ì¸ë±ìŠ¤: {min_index})")

print()



# ===== 4. ì¡°ê±´ë¶€ í•„í„°ë§ =====

print("=== 4. ì¡°ê±´ë¶€ í•„í„°ë§ ===")



# 4-1. ì¡°ê±´ìœ¼ë¡œ ë°ì´í„° ì„ íƒ

print("ğŸ“Š 4-1. ìœ„í—˜ ê±°ë¦¬ ê°ì§€ (5m ì´í•˜)")

dangerous_mask = all_distances <= 5.0  # ì¡°ê±´ ë§ˆìŠ¤í¬ ìƒì„±

print("ì¡°ê±´ ë§ˆìŠ¤í¬:", dangerous_mask)



dangerous_distances = all_distances[dangerous_mask]  # ì¡°ê±´ì— ë§ëŠ” ê°’ë“¤

safe_distances = all_distances[all_distances > 5.0]  # í•œ ì¤„ë¡œ í‘œí˜„



print("ìœ„í—˜ ê±°ë¦¬:", dangerous_distances)

print("ì•ˆì „ ê±°ë¦¬:", safe_distances)

print("ìœ„í—˜ ì„¼ì„œ ê°œìˆ˜:", len(dangerous_distances))

print("ì•ˆì „ ì„¼ì„œ ê°œìˆ˜:", len(safe_distances))

print()



# 4-2. ë³µí•© ì¡°ê±´

print("ğŸ“Š 4-2. ë³µí•© ì¡°ê±´ (5m ì´í•˜ OR 15m ì´ìƒ)")

extreme_distances = all_distances[(all_distances <= 5.0) | (all_distances >= 15.0)]

print("ê·¹í•œ ê±°ë¦¬ (ë„ˆë¬´ ê°€ê¹ê±°ë‚˜ ë„ˆë¬´ ë¨¼):", extreme_distances)



normal_distances = all_distances[(all_distances > 5.0) & (all_distances < 15.0)]

print("ì •ìƒ ë²”ìœ„ ê±°ë¦¬:", normal_distances)

print()



# 5-1. 2D ì„¼ì„œ ë°°ì—´ ìƒì„±

print("ğŸ“Š 5-1. ë‹¤ë°©í–¥ ì„¼ì„œ ë§¤íŠ¸ë¦­ìŠ¤")

# 4x4 ì„¼ì„œ ê·¸ë¦¬ë“œ (ì „ë°©, ìš°ì¸¡, í›„ë°©, ì¢Œì¸¡)

sensor_matrix = np.array([

    [8.5, 12.3, 15.7, 11.2],  # ì „ë°© 4ê°œ ì„¼ì„œ

    [9.1, 6.8, 13.4, 7.5],    # ìš°ì¸¡ 4ê°œ ì„¼ì„œ

    [10.2, 8.9, 14.1, 9.7],   # í›„ë°© 4ê°œ ì„¼ì„œ

    [7.3, 11.6, 12.8, 6.4]    # ì¢Œì¸¡ 4ê°œ ì„¼ì„œ

])



print("ì„¼ì„œ ë§¤íŠ¸ë¦­ìŠ¤:")

print(sensor_matrix)

print("ë§¤íŠ¸ë¦­ìŠ¤ ëª¨ì–‘:", sensor_matrix.shape)  # (í–‰, ì—´)

print("ì´ ì„¼ì„œ ê°œìˆ˜:", sensor_matrix.size)

print()



# 5-2. í–‰/ì—´ë³„ ì—°ì‚°

print("ğŸ“Š 5-2. ë°©í–¥ë³„ ì„¼ì„œ ë¶„ì„")

print("ì „ë°© ì„¼ì„œ (ì²« ë²ˆì§¸ í–‰):", sensor_matrix[0])

print("ì „ë°© ì„¼ì„œ í‰ê· :", np.mean(sensor_matrix[0]))



print("ëª¨ë“  ë°©í–¥ í‰ê· :")

direction_names = ['ì „ë°©', 'ìš°ì¸¡', 'í›„ë°©', 'ì¢Œì¸¡']

for i, direction in enumerate(direction_names):

    avg_distance = np.mean(sensor_matrix[i])

    print(f"  {direction}: {avg_distance:.2f}m")

print()



# 5-3. ì „ì²´ ë§¤íŠ¸ë¦­ìŠ¤ ì—°ì‚°

print("ğŸ“Š 5-3. ì „ì²´ ì„¼ì„œ í†µê³„")

print("ì „ì²´ í‰ê· :", np.mean(sensor_matrix))

print("ì „ì²´ ìµœëŒ€:", np.max(sensor_matrix))

print("ì „ì²´ ìµœì†Œ:", np.min(sensor_matrix))



# ê° ì—´ (ì„¼ì„œ ìœ„ì¹˜)ë³„ í‰ê· 

column_averages = np.mean(sensor_matrix, axis=0)  # axis=0: í–‰ ë°©í–¥ìœ¼ë¡œ í‰ê· 

print("ì„¼ì„œ ìœ„ì¹˜ë³„ í‰ê· :", column_averages)



# ê° í–‰ (ë°©í–¥)ë³„ í‰ê· 

row_averages = np.mean(sensor_matrix, axis=1)  # axis=1: ì—´ ë°©í–¥ìœ¼ë¡œ í‰ê· 

print("ë°©í–¥ë³„ í‰ê· :", row_averages)

print()



# ===== 6. GPS ì¢Œí‘œ ì²˜ë¦¬ =====

print("=== 6. GPS ì¢Œí‘œ ì²˜ë¦¬ ===")



# 6-1. ìœ„ì¹˜ ë²¡í„°

print("ğŸ“Š 6-1. í˜„ì¬ ìœ„ì¹˜ì™€ ëª©ì ì§€")

current_pos = np.array([127.123, 37.456])  # [ê²½ë„, ìœ„ë„]

destination = np.array([127.145, 37.478])



print("í˜„ì¬ ìœ„ì¹˜:", current_pos)

print("ëª©ì ì§€:", destination)



# ì´ë™ ë²¡í„° ê³„ì‚°

movement_vector = destination - current_pos

print("ì´ë™ ë²¡í„°:", movement_vector)

print("ê²½ë„ ì´ë™:", movement_vector[0])

print("ìœ„ë„ ì´ë™:", movement_vector[1])

print()



# 6-2. ê±°ë¦¬ ê³„ì‚°

print("ğŸ“Š 6-2. ì§ì„  ê±°ë¦¬ ê³„ì‚°")

# ìœ í´ë¦¬ë“œ ê±°ë¦¬ ê³µì‹: sqrt((x2-x1)^2 + (y2-y1)^2)

distance = np.sqrt(np.sum(movement_vector**2))

print("ì§ì„  ê±°ë¦¬:", distance)



# ë§¨í•˜íƒ„ ê±°ë¦¬ (ê²©ì ë„ë¡œì—ì„œ ì‹¤ì œ ì´ë™ ê±°ë¦¬)

manhattan_distance = np.sum(np.abs(movement_vector))

print("ë§¨í•˜íƒ„ ê±°ë¦¬:", manhattan_distance)

print()



# ===== 7. ê²½ë¡œ ìƒì„± =====

print("=== 7. ê²½ë¡œ ìƒì„± ===")



# 7-1. ì§ì„  ê²½ë¡œ í¬ì¸íŠ¸ ìƒì„±

print("ğŸ“Š 7-1. ìë™ ê²½ë¡œ ìƒì„±")

num_waypoints = 8

route_x = np.linspace(current_pos[0], destination[0], num_waypoints)

route_y = np.linspace(current_pos[1], destination[1], num_waypoints)



print("ê²½ë¡œ X ì¢Œí‘œ:", route_x)

print("ê²½ë¡œ Y ì¢Œí‘œ:", route_y)



# ì¢Œí‘œë¥¼ í•˜ë‚˜ì˜ 2D ë°°ì—´ë¡œ ê²°í•©

route_points = np.column_stack((route_x, route_y))

print("ê²½ë¡œ í¬ì¸íŠ¸ ë§¤íŠ¸ë¦­ìŠ¤:")

print(route_points)

print("ê²½ë¡œ ëª¨ì–‘:", route_points.shape)

print()



# 7-2. ê° êµ¬ê°„ë³„ ê±°ë¦¬ ê³„ì‚°

print("ğŸ“Š 7-2. êµ¬ê°„ë³„ ì´ë™ ê±°ë¦¬")

# ì—°ì†ëœ í¬ì¸íŠ¸ ê°„ ê±°ë¦¬ ê³„ì‚°

segment_distances = []

for i in range(len(route_points) - 1):

    point1 = route_points[i]

    point2 = route_points[i + 1]

    segment_dist = np.sqrt(np.sum((point2 - point1)**2))

    segment_distances.append(segment_dist)

    print(f"êµ¬ê°„ {i+1}: {segment_dist:.6f}")



total_route_distance = np.sum(segment_distances)

print(f"ì´ ê²½ë¡œ ê±°ë¦¬: {total_route_distance:.6f}")

print()



# ===== 8. ì‹¤ì „ ì‘ìš© - ì¥ì• ë¬¼ íšŒí”¼ =====

print("=== 8. ì‹¤ì „ ì‘ìš© - ì¥ì• ë¬¼ íšŒí”¼ ===")



# 8-1. ì¥ì• ë¬¼ ìœ„ì¹˜

print("ğŸ“Š 8-1. ì¥ì• ë¬¼ ê°ì§€ ë° ë¶„ì„")

obstacles = np.array([

    [127.130, 37.460],  # ì¥ì• ë¬¼ 1

    [127.135, 37.465],  # ì¥ì• ë¬¼ 2

    [127.140, 37.470]   # ì¥ì• ë¬¼ 3

])



print("ê°ì§€ëœ ì¥ì• ë¬¼ ìœ„ì¹˜:")

print(obstacles)



# í˜„ì¬ ìœ„ì¹˜ì—ì„œ ê° ì¥ì• ë¬¼ê¹Œì§€ì˜ ê±°ë¦¬

obstacle_distances = []

for i, obstacle in enumerate(obstacles):

    dist = np.sqrt(np.sum((obstacle - current_pos)**2))

    obstacle_distances.append(dist)

    print(f"ì¥ì• ë¬¼ {i+1}ê¹Œì§€ ê±°ë¦¬: {dist:.6f}")



# ê°€ì¥ ê°€ê¹Œìš´ ì¥ì• ë¬¼ ì°¾ê¸°

closest_obstacle_index = np.argmin(obstacle_distances)

closest_distance = obstacle_distances[closest_obstacle_index]

print(f"ê°€ì¥ ê°€ê¹Œìš´ ì¥ì• ë¬¼: {closest_obstacle_index+1}ë²ˆ (ê±°ë¦¬: {closest_distance:.6f})")

print()



# 8-2. ì•ˆì „ ë°˜ê²½ ì²´í¬

print("ğŸ“Š 8-2. ì•ˆì „ ë°˜ê²½ ì²´í¬")

safety_radius = 0.005  # ëŒ€ëµ 500m ì •ë„

dangerous_obstacles = []



for i, dist in enumerate(obstacle_distances):

    if dist <= safety_radius:

        dangerous_obstacles.append(i+1)

        print(f"âš ï¸ ì¥ì• ë¬¼ {i+1}: ìœ„í—˜ êµ¬ì—­ (ê±°ë¦¬: {dist:.6f})")

    else:

        print(f"âœ… ì¥ì• ë¬¼ {i+1}: ì•ˆì „ (ê±°ë¦¬: {dist:.6f})")



if dangerous_obstacles:

    print(f"ì´ {len(dangerous_obstacles)}ê°œ ì¥ì• ë¬¼ì´ ìœ„í—˜ êµ¬ì—­ì— ìˆìŠµë‹ˆë‹¤!")

else:

    print("ëª¨ë“  ì¥ì• ë¬¼ì´ ì•ˆì „ ê±°ë¦¬ì— ìˆìŠµë‹ˆë‹¤.")

print()

print("ğŸ“ NumPy ê¸°ì´ˆ í•™ìŠµ ì™„ë£Œ!")

print("ğŸ“š í•™ìŠµí•œ ë‚´ìš©:")

print("   âœ… ë°°ì—´ ìƒì„± ë° ê¸°ë³¸ ì—°ì‚°")

print("   âœ… í†µê³„ í•¨ìˆ˜ (í‰ê· , ìµœëŒ€, ìµœì†Œ ë“±)")

print("   âœ… ì¡°ê±´ë¶€ í•„í„°ë§ (ìœ„í—˜ ê°ì§€)")

print("   âœ… 2D ë°°ì—´ (ì„¼ì„œ ë§¤íŠ¸ë¦­ìŠ¤)")

print("   âœ… GPS ì¢Œí‘œ ì²˜ë¦¬")

print("   âœ… ê²½ë¡œ ìƒì„± ë° ê±°ë¦¬ ê³„ì‚°")

print("   âœ… ì¥ì• ë¬¼ ê°ì§€ ë° ì•ˆì „ ë°˜ê²½ ì²´í¬")

print("\nğŸš— ë‹¤ìŒì€ Pandasë¡œ ì£¼í–‰ ë°ì´í„°ë¥¼ ë¶„ì„í•´ë³´ê² ìŠµë‹ˆë‹¤!")

print("="*60)



```

```python
import numpy as np  # âœ… NumPy ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ npë¼ëŠ” ì´ë¦„ìœ¼ë¡œ ë¶ˆëŸ¬ì˜¤ê¸° (ê´€ë¡€ì ì¸ ì¶•ì•½ì–´)

# í˜„ì¬ ì„¤ì¹˜ëœ NumPy ë²„ì „ ì¶œë ¥
print("\U0001F697 ììœ¨ì£¼í–‰ NumPy ê¸°ì´ˆ í•™ìŠµ ì‹œì‘!")
print("\U0001F4E6 NumPy ë²„ì „:", np.__version__)
print()

# ================================================
# [1] NumPy ë°°ì—´ ê¸°ì´ˆ: ë°°ì—´ ìƒì„± ë° ë³€í™˜
# ================================================

print("=== 1. NumPy ë°°ì—´ ê¸°ì´ˆ ===")
print("\U0001F4CA 1-1. ì„¼ì„œ ê±°ë¦¬ ë°ì´í„°")

sensor_distances = [8.5, 12.3, 15.7, 2.1, 9.8]  # ì „ë°©ì— ìˆëŠ” ë¬¼ì²´ê¹Œì§€ì˜ ê±°ë¦¬ ì •ë³´
print("ì›ë³¸ ë¦¬ìŠ¤íŠ¸:", sensor_distances)
print("ë¦¬ìŠ¤íŠ¸ íƒ€ì…:", type(sensor_distances))  # <class 'list'> í˜•ì‹

np_distances = np.array(sensor_distances)  # ë¦¬ìŠ¤íŠ¸ë¥¼ NumPy ë°°ì—´ë¡œ ë³€í™˜
print("NumPy ë°°ì—´:", np_distances)
print("NumPy íƒ€ì…:", type(np_distances))       # <class 'numpy.ndarray'>
print("ë°°ì—´ ëª¨ì–‘:", np_distances.shape)        # (5,) â†’ 1ì°¨ì› ë°°ì—´
print("ë°°ì—´ í¬ê¸°:", np_distances.size)         # ì›ì†Œ ê°œìˆ˜
print()

print("\U0001F4CA 1-2. ë‹¤ì–‘í•œ ë°°ì—´ ìƒì„± ë°©ë²•")

zeros_array = np.zeros(5)         # 0ìœ¼ë¡œ ì±„ì›Œì§„ ë°°ì—´ (ê¸°ë³¸ê°’ ì´ˆê¸°í™”)
ones_array = np.ones(4)           # 1ë¡œ ì±„ì›Œì§„ ë°°ì—´ (ê¸°ë³¸ ë‹¨ìœ„ê°’ ì„¸íŒ…)
range_array = np.arange(0, 10, 2) # 0ë¶€í„° 10 ë¯¸ë§Œê¹Œì§€ 2 ê°„ê²©ìœ¼ë¡œ ìƒì„±ëœ ë°°ì—´
linspace_array = np.linspace(0, 100, 5)  # 0ë¶€í„° 100ê¹Œì§€ë¥¼ ê· ë“±í•˜ê²Œ 5ê°œë¡œ ë‚˜ëˆˆ ë°°ì—´

print("0ìœ¼ë¡œ ì±„ìš´ ë°°ì—´:", zeros_array)
print("1ë¡œ ì±„ìš´ ë°°ì—´:", ones_array)
print("ë²”ìœ„ ë°°ì—´ (0~10, 2ì”©):", range_array)
print("êµ¬ê°„ ë°°ì—´ (0~100, 5ê°œ):", linspace_array)
print()

# ================================================
# [2] ë°°ì—´ ì—°ì‚°: ê¸°ë³¸ ì‚°ìˆ  + ë°°ì—´ ê°„ ì—°ì‚°
# ================================================

print("=== 2. ë°°ì—´ ì—°ì‚° ===")
print("\U0001F4CA 2-1. ì„¼ì„œ ë°ì´í„° ê³„ì‚°")

distances = np.array([8.5, 12.3, 15.7, 2.1, 9.8])  # ì‹¤ì œ ê±°ë¦¬ ì„¼ì„œ ë°ì´í„° ë°°ì—´
print("ì›ë³¸ ê±°ë¦¬:", distances)

distances_in_cm = distances * 100  # ë¯¸í„°ë¥¼ ì„¼í‹°ë¯¸í„°ë¡œ í™˜ì‚° (ë°°ì—´ ë‚´ ëª¨ë“  ìš”ì†Œì— 100 ê³±í•¨)
distances_plus_offset = distances + 1.5  # ëª¨ë“  ì„¼ì„œ ê±°ë¦¬ê°’ì— 1.5m ì˜¤í”„ì…‹ ì¶”ê°€ (ë³´ì •)

print("ì„¼í‹°ë¯¸í„° ë³€í™˜:", distances_in_cm)
print("ì˜¤í”„ì…‹ ì ìš©:", distances_plus_offset)
print()

print("\U0001F4CA 2-2. ë‘ ì„¼ì„œ ë°ì´í„° ë¹„êµ")

front_sensors = np.array([8.5, 12.3, 15.7])  # ì „ë°© ì„¼ì„œ ê±°ë¦¬
rear_sensors = np.array([6.2, 9.1, 13.4])    # í›„ë°© ì„¼ì„œ ê±°ë¦¬

print("ì „ë°© ì„¼ì„œ:", front_sensors)
print("í›„ë°© ì„¼ì„œ:", rear_sensors)

print("ì„¼ì„œ ì°¨ì´:", front_sensors - rear_sensors)  # ê° ìœ„ì¹˜ë³„ ì „ë°©ê³¼ í›„ë°© ê±°ë¦¬ ì°¨ì´ ê³„ì‚°
print("í‰ê·  ê±°ë¦¬:", (front_sensors + rear_sensors) / 2)  # ê° ìœ„ì¹˜ë³„ ì „í›„ í‰ê·  ê±°ë¦¬ ê³„ì‚°
print()

# ================================================
# [3] ë°°ì—´ í†µê³„ í•¨ìˆ˜: ìµœëŒ€, ìµœì†Œ, í‰ê·  ë“±
# ================================================

print("=== 3. ë°°ì—´ í†µê³„ í•¨ìˆ˜ ===")
print("\U0001F4CA 3-1. ì„¼ì„œ ë°ì´í„° í†µê³„")

all_distances = np.array([8.5, 12.3, 15.7, 2.1, 9.8, 6.2, 11.4, 3.9])  # ëª¨ë“  ì„¼ì„œ ê±°ë¦¬ ë°ì´í„° í†µí•©
print("ì „ì²´ ì„¼ì„œ ë°ì´í„°:", all_distances)

print("ìµœëŒ€ê°’:", np.max(all_distances))        # ê°€ì¥ ë¨¼ ê±°ë¦¬
print("ìµœì†Œê°’:", np.min(all_distances))        # ê°€ì¥ ê°€ê¹Œìš´ ê±°ë¦¬
print("í‰ê· ê°’:", np.mean(all_distances))      # ì „ì²´ ê±°ë¦¬ì˜ í‰ê· 
print("ì¤‘ê°„ê°’:", np.median(all_distances))    # ì¤‘ì•™ê°’ (ì •ë ¬ í›„ ê°€ìš´ë° ê°’)
print("í‘œì¤€í¸ì°¨:", np.std(all_distances))     # ê±°ë¦¬ê°’ë“¤ì˜ ë¶„ì‚° ì •ë„
print("í•©ê³„:", np.sum(all_distances))          # ì „ì²´ ê±°ë¦¬ì˜ í•©ê³„
print()

print("\U0001F4CA 3-2. íŠ¹ì • ê°’ì˜ ìœ„ì¹˜ ì°¾ê¸°")

max_index = np.argmax(all_distances)  # ìµœëŒ€ê°’ì˜ ì¸ë±ìŠ¤ ìœ„ì¹˜ ë°˜í™˜
min_index = np.argmin(all_distances)  # ìµœì†Œê°’ì˜ ì¸ë±ìŠ¤ ìœ„ì¹˜ ë°˜í™˜

print(f"ê°€ì¥ ë¨¼ ê±°ë¦¬: {all_distances[max_index]}m (ì¸ë±ìŠ¤: {max_index})")
print(f"ê°€ì¥ ê°€ê¹Œìš´ ê±°ë¦¬: {all_distances[min_index]}m (ì¸ë±ìŠ¤: {min_index})")
print()

# ================================================
# [4] ì¡°ê±´ë¶€ í•„í„°ë§: ìœ„í—˜/ì•ˆì „ ê±°ë¦¬ ë¶„ë¥˜
# ================================================

print("=== 4. ì¡°ê±´ë¶€ í•„í„°ë§ ===")
print("\U0001F4CA 4-1. ìœ„í—˜ ê±°ë¦¬ ê°ì§€ (5m ì´í•˜)")

dangerous_mask = all_distances <= 5.0  # 5m ì´í•˜ì¼ ê²½ìš° Trueê°€ ë˜ëŠ” ë§ˆìŠ¤í¬ ìƒì„±
print("ì¡°ê±´ ë§ˆìŠ¤í¬:", dangerous_mask)   # Boolean ë°°ì—´ë¡œ ì¡°ê±´ ë§Œì¡± ì—¬ë¶€ ì¶œë ¥

dangerous_distances = all_distances[dangerous_mask]  # ë§ˆìŠ¤í¬ë¥¼ ì´ìš©í•´ ìœ„í—˜ ê±°ë¦¬ ì¶”ì¶œ
safe_distances = all_distances[all_distances > 5.0]   # 5m ì´ˆê³¼ì¸ ì•ˆì „ ê±°ë¦¬ ì¶”ì¶œ

print("ìœ„í—˜ ê±°ë¦¬:", dangerous_distances)
print("ì•ˆì „ ê±°ë¦¬:", safe_distances)
print("ìœ„í—˜ ì„¼ì„œ ê°œìˆ˜:", len(dangerous_distances))
print("ì•ˆì „ ì„¼ì„œ ê°œìˆ˜:", len(safe_distances))
print()

print("\U0001F4CA 4-2. ë³µí•© ì¡°ê±´ (5m ì´í•˜ OR 15m ì´ìƒ)")

# ë…¼ë¦¬ ì—°ì‚°ì | (or), & (and)ë¥¼ ì‚¬ìš©í•œ ë³µí•© ì¡°ê±´ í•„í„°ë§
# ë„ˆë¬´ ê°€ê¹Œìš´ ê±°ë¦¬ (5m ì´í•˜) ë˜ëŠ” ë„ˆë¬´ ë¨¼ ê±°ë¦¬ (15m ì´ìƒ) â†’ ê·¹í•œ ê±°ë¦¬ë¡œ ê°„ì£¼
extreme_distances = all_distances[(all_distances <= 5.0) | (all_distances >= 15.0)]

# ì ì ˆí•œ ê±°ë¦¬ ë²”ìœ„ (5m ì´ˆê³¼ & 15m ë¯¸ë§Œ) â†’ ì •ìƒì ì¸ ì£¼í–‰ ê°€ëŠ¥ êµ¬ê°„
normal_distances = all_distances[(all_distances > 5.0) & (all_distances < 15.0)]

print("ê·¹í•œ ê±°ë¦¬ (ë„ˆë¬´ ê°€ê¹ê±°ë‚˜ ë„ˆë¬´ ë¨¼):", extreme_distances)
print("ì •ìƒ ë²”ìœ„ ê±°ë¦¬:", normal_distances)
print()

# ================================================
# [5] 2D ë°°ì—´: ë‹¤ë°©í–¥ ì„¼ì„œ ë°ì´í„° ì²˜ë¦¬
# ================================================

print("ğŸ“Š 5-1. ë‹¤ë°©í–¥ ì„¼ì„œ ë§¤íŠ¸ë¦­ìŠ¤")

# ììœ¨ì£¼í–‰ì°¨ì˜ 4ë°©í–¥(ì „ë°©, ìš°ì¸¡, í›„ë°©, ì¢Œì¸¡) ì„¼ì„œë¡œë¶€í„° ìˆ˜ì§‘ëœ ê±°ë¦¬ ë°ì´í„°ë¥¼ 2ì°¨ì› ë°°ì—´ë¡œ í‘œí˜„
sensor_matrix = np.array([
    [8.5, 12.3, 15.7, 11.2],  # ì „ë°© ì„¼ì„œ (4ê°œ)
    [9.1, 6.8, 13.4, 7.5],    # ìš°ì¸¡ ì„¼ì„œ (4ê°œ)
    [10.2, 8.9, 14.1, 9.7],   # í›„ë°© ì„¼ì„œ (4ê°œ)
    [7.3, 11.6, 12.8, 6.4]    # ì¢Œì¸¡ ì„¼ì„œ (4ê°œ)
])

print("ì„¼ì„œ ë§¤íŠ¸ë¦­ìŠ¤:")
print(sensor_matrix)
print("ë§¤íŠ¸ë¦­ìŠ¤ ëª¨ì–‘:", sensor_matrix.shape)  # (4í–‰, 4ì—´) â†’ 4ë°©í–¥ * 4ì„¼ì„œ
print("ì´ ì„¼ì„œ ê°œìˆ˜:", sensor_matrix.size)   # ì „ì²´ ì›ì†Œ ê°œìˆ˜ (16ê°œ)
print()

print("ğŸ“Š 5-2. ë°©í–¥ë³„ ì„¼ì„œ ë¶„ì„")

# ì „ë°©(ì²« ë²ˆì§¸ í–‰)ì˜ ì„¼ì„œ ê°’ë§Œ ì¶”ì¶œí•˜ì—¬ í‰ê· ì„ êµ¬í•¨
print("ì „ë°© ì„¼ì„œ (ì²« ë²ˆì§¸ í–‰):", sensor_matrix[0])
print("ì „ë°© ì„¼ì„œ í‰ê· :", np.mean(sensor_matrix[0]))

# ì „ì²´ ë°©í–¥ë³„ í‰ê·  ì¶œë ¥ (forë¬¸ í™œìš©)
direction_names = ['ì „ë°©', 'ìš°ì¸¡', 'í›„ë°©', 'ì¢Œì¸¡']
print("ëª¨ë“  ë°©í–¥ í‰ê· :")
for i, direction in enumerate(direction_names):
    avg_distance = np.mean(sensor_matrix[i])
    print(f"  {direction}: {avg_distance:.2f}m")
print()

print("ğŸ“Š 5-3. ì „ì²´ ì„¼ì„œ í†µê³„")

# ì „ì²´ ì„¼ì„œ ë°°ì—´ì— ëŒ€í•œ í†µê³„
print("ì „ì²´ í‰ê· :", np.mean(sensor_matrix))
print("ì „ì²´ ìµœëŒ€:", np.max(sensor_matrix))
print("ì „ì²´ ìµœì†Œ:", np.min(sensor_matrix))

# ì—´ ë°©í–¥ í‰ê·  (ê° ìœ„ì¹˜ë³„ ê±°ë¦¬ í‰ê· )
column_averages = np.mean(sensor_matrix, axis=0)  # í–‰ì„ ë”°ë¼ ì—´ë³„ í‰ê· 
print("ì„¼ì„œ ìœ„ì¹˜ë³„ í‰ê· :", column_averages)

# í–‰ ë°©í–¥ í‰ê·  (ê° ë°©í–¥ë³„ í‰ê· )
row_averages = np.mean(sensor_matrix, axis=1)  # ì—´ì„ ë”°ë¼ í–‰ë³„ í‰ê· 
print("ë°©í–¥ë³„ í‰ê· :", row_averages)
print()

# ================================================
# [6] GPS ì¢Œí‘œ ì²˜ë¦¬: ìœ„ì¹˜ ë²¡í„° ë° ê±°ë¦¬ ê³„ì‚°
# ================================================

print("=== 6. GPS ì¢Œí‘œ ì²˜ë¦¬ ===")
print("ğŸ“Š 6-1. í˜„ì¬ ìœ„ì¹˜ì™€ ëª©ì ì§€")

# í˜„ì¬ ìœ„ì¹˜ì™€ ëª©ì ì§€ GPS ì¢Œí‘œ (ê²½ë„, ìœ„ë„ ìˆœì„œ)
current_pos = np.array([127.123, 37.456])
destination = np.array([127.145, 37.478])

print("í˜„ì¬ ìœ„ì¹˜:", current_pos)
print("ëª©ì ì§€:", destination)

# ë‘ ì§€ì  ê°„ ë²¡í„° ì°¨ì´ë¥¼ í†µí•´ ì´ë™ ë²¡í„° ê³„ì‚°
movement_vector = destination - current_pos

print("ì´ë™ ë²¡í„°:", movement_vector)
print("ê²½ë„ ì´ë™:", movement_vector[0])
print("ìœ„ë„ ì´ë™:", movement_vector[1])
print()

print("ğŸ“Š 6-2. ì§ì„  ê±°ë¦¬ ê³„ì‚°")

# ìœ í´ë¦¬ë“œ ê±°ë¦¬ ê³„ì‚° (í”¼íƒ€ê³ ë¼ìŠ¤ ì •ë¦¬ ê¸°ë°˜)
distance = np.sqrt(np.sum(movement_vector**2))
print("ì§ì„  ê±°ë¦¬:", distance)

# ë§¨í•´íŠ¼ ê±°ë¦¬ ê³„ì‚° (ë„ì‹œ ë¸”ë¡ì²˜ëŸ¼ ê²½ë¡œë¥¼ êº¾ì–´ì„œ ì´ë™í•œë‹¤ê³  ê°€ì •)
manhattan_distance = np.sum(np.abs(movement_vector))
print("ë§¨í•˜íƒ„ ê±°ë¦¬:", manhattan_distance)
print()

# ================================================
# [7] ê²½ë¡œ ìƒì„±: ìë™ waypoint ë° ê±°ë¦¬ ê³„ì‚°
# ================================================

print("=== 7. ê²½ë¡œ ìƒì„± ===")
print("ğŸ“Š 7-1. ìë™ ê²½ë¡œ ìƒì„±")

# ëª©ì ì§€ê¹Œì§€ ì´ ëª‡ ê°œì˜ ê²½ìœ ì§€(waypoints)ë¥¼ ìƒì„±í• ì§€ ì„¤ì •
num_waypoints = 8

# ê²½ë„ì™€ ìœ„ë„ì— ëŒ€í•´ ì„ í˜• ë³´ê°„ ìˆ˜í–‰ (np.linspace)
route_x = np.linspace(current_pos[0], destination[0], num_waypoints)
route_y = np.linspace(current_pos[1], destination[1], num_waypoints)

print("ê²½ë¡œ X ì¢Œí‘œ:", route_x)
print("ê²½ë¡œ Y ì¢Œí‘œ:", route_y)

# ê²½ë¡œë¥¼ (ê²½ë„, ìœ„ë„) ìŒìœ¼ë¡œ ë¬¶ì–´ 2ì°¨ì› ë°°ì—´ë¡œ ê²°í•©
route_points = np.column_stack((route_x, route_y))

print("ê²½ë¡œ í¬ì¸íŠ¸ ë§¤íŠ¸ë¦­ìŠ¤:")
print(route_points)
print("ê²½ë¡œ ëª¨ì–‘:", route_points.shape)
print()

print("ğŸ“Š 7-2. êµ¬ê°„ë³„ ì´ë™ ê±°ë¦¬")

# ì—°ì†ëœ í¬ì¸íŠ¸ ê°„ ê±°ë¦¬ ê³„ì‚° (ìœ í´ë¦¬ë“œ ê±°ë¦¬ ê¸°ì¤€)
segment_distances = []
for i in range(len(route_points) - 1):
    point1 = route_points[i]
    point2 = route_points[i + 1]
    segment_dist = np.sqrt(np.sum((point2 - point1)**2))
    segment_distances.append(segment_dist)
    print(f"êµ¬ê°„ {i+1}: {segment_dist:.6f}")

total_route_distance = np.sum(segment_distances)
print(f"ì´ ê²½ë¡œ ê±°ë¦¬: {total_route_distance:.6f}")
print()

# ================================================
# [8] ì‹¤ì „ ì‘ìš© - ì¥ì• ë¬¼ íšŒí”¼ ì‹œë®¬ë ˆì´ì…˜
# ================================================

print("=== 8. ì‹¤ì „ ì‘ìš© - ì¥ì• ë¬¼ íšŒí”¼ ===")
print("ğŸ“Š 8-1. ì¥ì• ë¬¼ ê°ì§€ ë° ë¶„ì„")

# ììœ¨ì£¼í–‰ì°¨ ê·¼ì²˜ì— ìˆëŠ” ì¥ì• ë¬¼ GPS ì¢Œí‘œ
obstacles = np.array([
    [127.130, 37.460],  # ì¥ì• ë¬¼ 1
    [127.135, 37.465],  # ì¥ì• ë¬¼ 2
    [127.140, 37.470]   # ì¥ì• ë¬¼ 3
])

print("ê°ì§€ëœ ì¥ì• ë¬¼ ìœ„ì¹˜:")
print(obstacles)

# ê° ì¥ì• ë¬¼ê¹Œì§€ì˜ ê±°ë¦¬ ê³„ì‚°
obstacle_distances = []
for i, obstacle in enumerate(obstacles):
    dist = np.sqrt(np.sum((obstacle - current_pos)**2))  # ìœ í´ë¦¬ë“œ ê±°ë¦¬
    obstacle_distances.append(dist)
    print(f"ì¥ì• ë¬¼ {i+1}ê¹Œì§€ ê±°ë¦¬: {dist:.6f}")

# ê°€ì¥ ê°€ê¹Œìš´ ì¥ì• ë¬¼ ì‹ë³„
closest_obstacle_index = np.argmin(obstacle_distances)
closest_distance = obstacle_distances[closest_obstacle_index]
print(f"ê°€ì¥ ê°€ê¹Œìš´ ì¥ì• ë¬¼: {closest_obstacle_index+1}ë²ˆ (ê±°ë¦¬: {closest_distance:.6f})")
print()

print("ğŸ“Š 8-2. ì•ˆì „ ë°˜ê²½ ì²´í¬")

# ì°¨ëŸ‰ ê¸°ì¤€ìœ¼ë¡œ ì„¤ì •í•œ ìœ„í—˜ ë°˜ê²½ (ë‹¨ìœ„: ì•½ 0.005ë„ â‰’ 500m)
safety_radius = 0.005

# ìœ„í—˜ ë°˜ê²½ ë‚´ ì¥ì• ë¬¼ ì—¬ë¶€ íŒë³„
dangerous_obstacles = []
for i, dist in enumerate(obstacle_distances):
    if dist <= safety_radius:
        dangerous_obstacles.append(i+1)
        print(f"âš ï¸ ì¥ì• ë¬¼ {i+1}: ìœ„í—˜ êµ¬ì—­ (ê±°ë¦¬: {dist:.6f})")
    else:
        print(f"âœ… ì¥ì• ë¬¼ {i+1}: ì•ˆì „ (ê±°ë¦¬: {dist:.6f})")

# ê²°ê³¼ ìš”ì•½ ì¶œë ¥
if dangerous_obstacles:
    print(f"ì´ {len(dangerous_obstacles)}ê°œ ì¥ì• ë¬¼ì´ ìœ„í—˜ êµ¬ì—­ì— ìˆìŠµë‹ˆë‹¤!")
else:
    print("ëª¨ë“  ì¥ì• ë¬¼ì´ ì•ˆì „ ê±°ë¦¬ì— ìˆìŠµë‹ˆë‹¤.")
print()

print("ğŸ“ NumPy ê¸°ì´ˆ í•™ìŠµ ì™„ë£Œ!")

print("ğŸ“š í•™ìŠµí•œ ë‚´ìš©:")

print("   âœ… ë°°ì—´ ìƒì„± ë° ê¸°ë³¸ ì—°ì‚°")

print("   âœ… í†µê³„ í•¨ìˆ˜ (í‰ê· , ìµœëŒ€, ìµœì†Œ ë“±)")

print("   âœ… ì¡°ê±´ë¶€ í•„í„°ë§ (ìœ„í—˜ ê°ì§€)")

print("   âœ… 2D ë°°ì—´ (ì„¼ì„œ ë§¤íŠ¸ë¦­ìŠ¤)")

print("   âœ… GPS ì¢Œí‘œ ì²˜ë¦¬")

print("   âœ… ê²½ë¡œ ìƒì„± ë° ê±°ë¦¬ ê³„ì‚°")

print("   âœ… ì¥ì• ë¬¼ ê°ì§€ ë° ì•ˆì „ ë°˜ê²½ ì²´í¬")

print("\nğŸš— ë‹¤ìŒì€ Pandasë¡œ ì£¼í–‰ ë°ì´í„°ë¥¼ ë¶„ì„í•´ë³´ê² ìŠµë‹ˆë‹¤!")

print("="*60)
```
